```python
df_depsex = df.groupby('Department')['Sex'].value_counts(normalize=True).unstack().fillna(0)
```

```python
print (df_depsex)
```

```python
# GrÃ¡fico de barras para la distribuciÃ³n de gÃ©nero por fuente de departamento
fig1, ax1 = plt.subplots()
df_depsex.plot(kind='bar', stacked=True, ax=ax1)
ax1.set_title('DistribuciÃ³n de gÃ©nero por fuente de departamento')
ax1.set_xlabel('Departamento')
ax1.set_ylabel('Porcentaje')
ax1.legend(title='Sexo')
```

```python
# Contar el nÃºmero de personas contratadas desde cada fuente de reclutamiento
recruitment_source_counts = df['RecruitmentSource'].value_counts()

# Plot the number of people hired from each recruitment source
plt.figure(figsize=(10, 6))
recruitment_source_counts.plot(kind='bar', color='skyblue')
plt.title('Number of People Hired from Each Recruitment Source')
plt.xlabel('Recruitment Source')
plt.ylabel('Number of Hires')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python

# Para visualizar la distribuciÃ³n actual de la diversidad en la compaÃ±Ã­a
# Histograma de GenderID
plt.subplot(1, 3, 1)
df['Sex'].value_counts().plot(kind='bar')
plt.title('Sex')
plt.xlabel('Sex')
plt.ylabel('Count')

# Histograma de RaceDesc
plt.subplot(1, 3, 3)
df['RaceDesc'].value_counts().plot(kind='bar')
plt.title('Race')
plt.xlabel('Race')
plt.ylabel('Count')

# Mostrar los grÃ¡ficos
plt.tight_layout()
plt.show()
```

```python
aux = df.copy()
orden_custom_filtrado =   ["On-line Web application","Employee Referral", "Website", "Other", "Diversity Job Fair","Indeed", "LinkedIn","Google Search", "CareerBuilder"]
aux['RecruitmentSource'] = pd.Categorical(aux['RecruitmentSource'], orden_custom_filtrado )
df_genderbysource = aux.groupby('RecruitmentSource')['Sex'].value_counts(normalize=True).unstack().fillna(0)
```
